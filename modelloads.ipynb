```python
import tensorflow as tf
import os
import cv2
import imghdr
import numpy as np
from matplotlib import pyplot as plt
from tensorflow import keras
from tensorflow.keras.models import load_model
```

```python
#Model loads ekranı
mil_model = load_model(os.path.join('models','milcropeed.h5'))
mesh_model = load_model(os.path.join('models','mesh_40_cropped.h5'))
pres_model = load_model(os.path.join('models','pres_cropped40epoch.h5'))
hiz_model = load_model(os.path.join('models','hizfull2.h5'))
```

```python
img1 = cv2.imread('9485_or_r.jpg')
```

```python
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
resize1 = tf.image.resize(img1, (149,560))
plt.imshow(resize1.numpy().astype(int))
plt.show()
```

```python
# Önce görüntüyü normalize edin ve yeniden boyutlandırın
normalized_resized_image0 = resize1/255  # Örneğin, 0 ile 255 arasındaki piksel değerlerini [0, 1] aralığına normalize ediyoruz
input_image1 = np.expand_dims(normalized_resized_image0, 0)  # Girişi tek bir örnek olarak ekliyoruz

# Şimdi modelinize tahmin yaptırabilirsiniz
hiz_new = hiz_model.predict(input_image1)

# Tahmin edilen sınıfı alın
predicted_class_0 = np.argmax(hiz_new)

# Tahmin edilen sınıfın indeksini ve en yüksek olasılığı ekrana yazdırın
print(f"Predicted Class Index: {predicted_class_0}")
print(f"Highest Probability: {hiz_new[0, predicted_class_0]}")

#class 0 = hiz 30m/dk
#class 1 = hiz 40m/dk
#class 2 = hiz 50m/dk

# Eğer sınıf etiketleriniz varsa, sınıf etiketini de alabilirsiniz
class_labels0 = ["Class 0: hiz 30m/dk", "Class 1: hiz 40m/dk", "Class 2: hiz 50m/dk"]  # Örnek sınıf etiketleri
predicted_class_label0 = class_labels0[predicted_class_0]
print(hiz_new)
print(f"Predicted Class: {predicted_class_label0}")
```

```python
img = cv2.imread('resized_img.png')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
resize = tf.image.resize(img, (148,148))
plt.imshow(resize.numpy().astype(int))
plt.show()
```

```python
# Önce görüntüyü normalize edin ve yeniden boyutlandırın
normalized_resized_image2 = resize/255  # Örneğin, 0 ile 255 arasındaki piksel değerlerini [0, 1] aralığına normalize ediyoruz
input_image2 = np.expand_dims(normalized_resized_image2, 0)  # Girişi tek bir örnek olarak ekliyoruz

# Şimdi modelinize tahmin yaptırabilirsiniz
mil_model = mil_model.predict(input_image2)

# Tahmin edilen sınıfı alın
predicted_class_2 = np.argmax(mil_model)

# Tahmin edilen sınıfın indeksini ve en yüksek olasılığı ekrana yazdırın
print(f"Predicted Class Index: {predicted_class_2}")
print(f"Highest Probability: {mil_model[0, predicted_class_2]}")

#class 0 = 12 mm 
#class 1 = 16 mm


# Eğer sınıf etiketleriniz varsa, sınıf etiketini de alabilirsiniz
class_labels = ["Class 0: 12 mm", "Class 1: 16 mm"]  # Örnek sınıf etiketleri
predicted_class_label_2 = class_labels[predicted_class_2]
print(mil_model)
print(f"Predicted Class: {predicted_class_label_2}")
```
